```python
# С использованием имеющихся данных построим таблицу для выполнения расчёта.

n11 = 42          # Будут голосовать за А в Курске ("успех")
n1d = 105         # Всего опрошенных в Курске
n12 = n1d - n11   # Будут голосовать не за А в Курске ("неудача")

n21 = 65          # Будут голосовать за А во Владивостоке ("успех")
n2d = 195         # Всего опрошенных во Владивостоке
n22 = n2d - n21   # Будут голосовать не за А во Владивостоке ("неудача")

nd1 = n11 + n21   # Всего "успешных" исходов
nd2 = n12 + n22   # Всего "неудачных" исходов
n = nd1 + nd2     # Всего исходов
nn = n1d + n2d    # Всего исходов (проверка, nn должно быть равно n)

if n == nn:
    print('Данные введены верно \n')
else:
    print('При вводе данных допущена ошибка, проверьте значения n11, nd1, n21, n2d \n')

# Вывод таблицы введённых и вычисленных значений
print('           ', 'Успех', '|', 'Неудача', '|', 'Всего')
print('Курск        ', n11, ' |   ', n12, '  | ', n1d)
print('Владивосток  ', n21, ' |  ', n22, '  | ', n2d)
print('Всего       ', nd1, ' |  ', nd2, '  | ', n)
```

```python
import math
from math import sqrt

# Вычисляем значение критерия Т
T = ((n11 / n1d) - (n21 / n2d)) / sqrt((nd1 / n) * (1 - nd1 / n) * (1 / n1d + 1 / n2d))
print('Значение критерия Т равно', T)
```

```python
import scipy
from scipy.stats import norm

alpha = 0.05
Tcrl = norm.ppf(alpha / 2)        # Критическое значение критерия слева
Tcrr = norm.ppf(1.0 - alpha / 2)  # Критическое значение критерия справа
print('Границы доверительного интервала: ', round(Tcrl, 2), 'и', round(Tcrr, 2))
if T < Tcrl or T > Tcrr:
    print('Значение критерия попало в критическую область')
else:
    print('Значение критерия попало в доверительную область')
```

```python
F = norm.cdf(T) - 0.5
print('Значение Ф(Т) равно', round(F, 3))

pval = min(2 * F, 2 - 2* F)
print('Значение pval равно', round(pval, 3))
```

```python
import numpy as np
import scipy

n = 10   # Количество горожан, принявших участие в эксперименте
m = 9    # Количество селян, принявших участие в эксперименте

# Вносим значения уровня давления испытуемых в виде массивов
city = np.array([130, 110, 120, 140, 200, 130, 140, 170, 160, 140])
vill = np.array([120, 190, 130, 160, 150, 120, 110, 120, 200])

# Объединяем два массива в единый для ранжирования элементов
cv = np.concatenate((city, vill), axis = 0)
print('Объединённый массив', cv)

# Вычисляем ранги элементов массива с учётом того, что значения в массиве могут повторяться
from scipy.stats import rankdata
ranks = rankdata(cv)
print('Ранги элементов    ', ranks)
```

```python
# Вычисляем значение W суммы рангов элементов выборки горожан 
W = 0
for i in range(n):
    W = W + ranks[i]
print('Значение W суммы рангов элементов выборки горожан равно', W)
```

```python
# Вычислим значения математического ожидания и дисперсии 
# элементов выборки, состоящей из рангов элементов выборки горожан

import math
from math import sqrt

M = n / 2 * (m + n + 1)         # Математическое ожидание
D = m * n / 12 * (m + n + 1)    # Дисперсия

# Критическое значение W, получаемое посредством центрирования и нормирования W
Wcr = (W - M) / sqrt(D)

print('Математическое ожидание', M) 
print('Дисперсия              ', D)
print('Критическое значение W ', round(Wcr, 3))
```

```python
import scipy
from scipy.stats import norm

alpha = 0.05
Wcrr = norm.ppf(1 - alpha)        # Критическое значение критерия справа
print('Граница доверительного интервала: ', round(Tcrl, 2))
if Wcr > Wcrr:
    print('Значение критерия попало в критическую область')
else:
    print('Значение критерия попало в доверительную область')
```

```python
F = norm.cdf(Wcr) - 0.5
print('Значение Ф(Т) равно', round(Wcr, 3))

pval = 1 - F
print('Значение pval равно', round(pval, 3))
```

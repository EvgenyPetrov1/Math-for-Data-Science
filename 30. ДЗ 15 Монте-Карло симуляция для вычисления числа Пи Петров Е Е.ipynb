```python
# Для нахождения значения числа Pi найдём площадь фигуры, 
# являющейся четвертью круга радиусом 1, и расположенной в том квадранте, 
# где значения X и Y больше нуля.

import random

print ('Вариант с четвертью круга радиусом 1 \n')

# Количество точек, попавших в область (x ** 2 + y ** 2 < 1.0)
val = 0.0

# Количество точек, используемых в начале эксперимента
quan = 100

# Проведём 6 симуляций для количества точек, бросаемого в область 1 х 1,
# увеличивая на каждом шаге число точек на порядок, чтобы оценить точность
for i in range (6):
    for i in range(quan):
        x = random.random()
        y = random.random()
        val += (x ** 2 + y ** 2 < 1.0)
    print ('Количество точек = ', quan, 
           'Вычисленное значение Pi = ', 4 * val / quan)
    val = 0.0
    quan = quan * 10
```

```python
# Рассмотрим второй вариант: с той частью шара радиусом 1, которая отсекается 
# координатными плоскостями, и где значения всех трёх координат положительны
    
print ('\nВариант с восьмой частью сферы радиусом 1 \n')    

# Количество точек, попавших в область (x ** 2 + y ** 2  + z ** 2 < 1.0)
val = 0.0

# Количество точек, используемых в начале эксперимента
quan = 100

# Проведём 6 симуляций для количества точек, бросаемого в область 1 х 1 х 1,
# увеличивая на каждом шаге число точек на порядок, чтобы оценить точность
for i in range (6):
    for i in range(quan):
        x = random.random()
        y = random.random()
        z = random.random()
        val += (x ** 2 + y ** 2 + z ** 2 < 1.0)
    print ('Количество точек = ', quan, 
           'Вычисленное значение Pi = ', 6 * val / quan)
    val = 0.0
    quan = quan * 10
```

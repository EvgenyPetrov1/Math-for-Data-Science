```python
import math
import numpy as np
import matplotlib.pyplot as plt

R = 0.85               # Начальное значение коэффициента корреляции
mu, sigma = 0, 2       # Математическое ожидание и среднеквадратичное отклонение
n = 1000               # Количество точек на графике

# Генерация нормально распределённых значений для первой выборки
N1 = np.random.normal(mu, sigma, n)

# Генерация нормально распределённых значений с учётом зависимости
# от коэффицента корреляции для второй выборки
N2 = R * N1 + math.sqrt(1 - R ** 2) * np.random.normal(mu, sigma, n)

# График
plt.scatter(x = N1, y = N2, alpha = 0.2)
```

```python
K = 0.3                # Начальное значение коэффициента силы зависимости
mu, sigma = 0, 2       # Математическое ожидание и среднеквадратичное отклонение
n = 500                # Количество точек на графике

# Генерация нормально распределённых значений для первой выборки
N1 = np.random.normal(mu, sigma, n)

# Генерация нормально распределённых значений с учётом
# коэффицента силы зависимости для второй выборки
N2 = K ** 1.2 * N1 ** 2 + K ** 1.4 * np.random.normal(mu, sigma, n)

# График
plt.scatter(x = N1, y = N2, alpha = 0.15)
```
